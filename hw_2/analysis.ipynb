```python
import os
fileList = [x[:-5] for x in os.popen('ls homework2_data/*.sam').readlines()]
for eachFile in fileList:
    os.system('samtools view -Sb {}.sam > {}.bam'.format(eachFile,eachFile))
    os.system('samtools sort -O bam -o {}.sorted.bam {}.bam'.format(eachFile,eachFile))
    os.system('htseq-count -f bam -t exon -i gene_name {}.sorted.bam homework2_data/genes.gtf 2>{}_htseq.log | grep -v [[:space:]]0 > {}_count.txt'.format(eachFile,eachFile,eachFile))
```

```python
import os
fileList = [x[:-1] for x in os.popen('ls homework2_data/*.txt').readlines()]
fileList
count={}
for eachFile in fileList:
    with open(eachFile,'r') as fileContent:
        for line in fileContent.readlines():
            if line[:2]=='__':
                continue
            line = line[:-1]
            res = line.split()
            count[res[0]]=count.get(res[0],0)+int(res[1])
print(count)
gene_name = count.keys()
gene_count = count.values()
with open('homework2_data/total_count.txt','w') as output:
    for i in range(len(gene_name)):
        output.write(gene_name[i]+'\t'+gene_count[i])            
```
